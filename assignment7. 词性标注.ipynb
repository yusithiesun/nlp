2、设计汉语的组合标注器，以1标注的结果计算其性能指标。

```python
#将文本转换为元组，方便使用nltk进行训练
#参考：https://www.jianshu.com/p/4d802e9bfb45
import thulac
import nltk
#f=open('Leipzig10ksents1.txt','r')
#text=f.readlines()
#print(text)
#把jupyter写崩了我也不知道为什么
with open('Leipzig10ksents1.txt','r') as f:
    s=f.readlines()

tagged_token=[]
for line in s:
    t=[]
    t.append([nltk.tag.str2tuple(t.replace('_','/')) for t in line.split()])
    tagged_token+=t
print(tagged_token[:100])
```

```python
#不使用正则表达式的词性标注
#划分训练集和测试集
size=int(len(tagged_token)*0.9)
train_set=tagged_token[:size]
test_set=tagged_token[size:]

t0=nltk.DefaultTagger('N')#默认规则：标注为名词
t1=nltk.UnigramTagger(train_set,backoff=t0)
#http://www.nltk.org/book/ch05.html
#Many words have been assigned a tag of None, because they were not among the 100 most frequent words. 
#In these cases we would like to assign the default tag of NN. 
#In other words, we want to use the lookup table first, and if it is unable to assign a tag, 
#then use the default tagger, a process known as backoff (5). 
#We do this by specifying one tagger as a parameter to the other, as shown below. 
#Now the lookup tagger will only store word-tag pairs for words other than nouns, 
#and whenever it cannot assign a tag to a word it will invoke the default tagger.
t2=nltk.BigramTagger(train_set,backoff=t1)

print(t2.evaluate(test_set))
```

```python
#使用正则表达式的词性标注
patterns=[(r'[0-9]+','M'),
          (r',|。|：|“|”|？|《|》|》……','W')]
dt=nltk.DefaultTagger('N')
rt=nltk.RegexpTagger(patterns,backoff=dt)
t1=nltk.UnigramTagger(train_set,backoff=rt)
t2=nltk.BigramTagger(train_set,backoff=t1)

print(t2.evaluate(test_set))
```
